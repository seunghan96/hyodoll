```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
scc_emergency_call = pd.read_csv('scc_emergency_call.csv', sep = ';')
```

```python
print(scc_emergency_call.shape)
print(scc_emergency_call.user_id.nunique())
scc_emergency_call.head()
```

```python
scc_emergency_call.head(30)
```

```python
scc_emergency_call['temp_val'] = 1
scc_emergency_call
```

```python


scc_emergency_call1 = pd.pivot_table(scc_emergency_call, values = 'temp_val', index = ['doll_id'],
                    columns = ['call_month'], aggfunc = np.sum,fill_value =0)

scc_emergency_call2 = pd.pivot_table(scc_emergency_call, values = 'temp_val', index = ['doll_id'],
                    columns = ['call_hour'], aggfunc = np.sum,fill_value =0)

scc_emergency_call1.columns = ['emergency_month'+str(col) for col in scc_emergency_call1.columns]
scc_emergency_call2.columns = ['emergency_hour'+str(col) for col in scc_emergency_call2.columns]
```

```python
scc_emergency_call3 = scc_emergency_call2.copy()
scc_emergency_call3['emergency_night']= scc_emergency_call3[['emergency_hour'+str(i) for i in [2,1,0,23,22,21]]].sum(axis =1)
scc_emergency_call3['emergency_evening']= scc_emergency_call3[['emergency_hour'+str(i) for i in [20,19,18,17]]].sum(axis =1)
scc_emergency_call3['emergency_afternoon']= scc_emergency_call3[['emergency_hour'+str(i) for i in [16,15,14,13,12]]].sum(axis =1)
scc_emergency_call3['emergency_morning']= scc_emergency_call3[['emergency_hour'+str(i) for i in [11,10,9,8,7]]].sum(axis =1)
scc_emergency_call3['emergency_dawn']= scc_emergency_call3[['emergency_hour'+str(i) for i in [6,5,4,3]]].sum(axis =1)
scc_emergency_call3 = scc_emergency_call3[['emergency_night',
                                         'emergency_evening',
                                         'emergency_afternoon',
                                         'emergency_morning',
                                         'emergency_dawn']]
```
