# 1. PCA visualization

```python
cluster_df = pd.read_csv('FA_with_cluster.csv')
cluster_df=cluster_df.set_index('Unnamed: 0')

col_list=list(cluster_df.columns[cluster_df.columns.str.contains('FA')])
```

```python
cluster_df
```

```python
cluster_df.groupby('cluster23').mean()
```

```python
def pca_n(n,df,cols):
    pca = PCA(n_components=n)
    pca.fit(df[cols].values)
    print(pca.explained_variance_ratio_,sum(pca.explained_variance_ratio_))
    
    #---------------------------------------------#
    pca_df=pd.DataFrame(pca.fit_transform(df[cols].values),index=df.index,columns=['pc'+str(i) for i in range(1,n+1)])
    pca_df=pd.concat([pca_df,df[df.columns[df.columns.str.contains('cluster')]]],axis=1)
    return pca_df
    
```

```python
pca_df=pca_n(3,cluster_df,col_list)
```

```python
cluster_cases=['22','23','33','43']

def pca_df_mxn(pca_df,cluster_case):
    cluster_case='cluster'+cluster_case
    cols = list(pca_df.columns[pca_df.columns.str.contains('pc')])
    cols2 = cols.copy()
    cols2.append(cluster_case)
    pca_df_mn=pca_df[cols2]
    print(pca_df_mn.groupby(cluster_case).mean())
    print('---------------------------------')
    print(pca_df_mn[cluster_case].value_counts())
    print('---------------------------------')
    viz_list = [['pc1','pc2'],['pc1','pc3'],['pc2','pc3']]
    for i, viz in enumerate(viz_list):
        plt.figure(i)
        colors=['red', 'green','blue','yellow','purple','black']
        sns.scatterplot(x=viz[0], y=viz[1], data=pca_df_mn, palette=colors,hue=cluster_case,s=8)

    cols3=col_list.copy()
    cols3.append(cluster_case)
    cluster_summary = cluster_df[cols3].groupby(cluster_case).mean().round(2)
    cluster_summary.columns=['새벽기상','식전약','정시-아침식사','식후약','정시-저녁식사','늦잠','장기이용','활동다양성','여름가을','종교인','차분취미','활발취미','지적취미','퀴즈','겨울','여름','봄']
    cluster_summary=cluster_summary[['새벽기상','늦잠','식전약','식후약','정시-아침식사','정시-저녁식사','장기이용','활동다양성','여름가을','종교인','차분취미',
    '활발취미','지적취미','퀴즈','겨울','여름','봄']]
    return cluster_summary
    
    
```

```python
cluster_cases=['22','23','33','43']

def pca_df_mxn2(pca_df,cluster_case):
    cluster_case='cluster'+cluster_case
    cols = list(pca_df.columns[pca_df.columns.str.contains('pc')])
    cols2 = cols.copy()
    cols2.append(cluster_case)
    pca_df_mn=pca_df[cols2]
    print(pca_df_mn.columns)
    print(pca_df_mn.groupby(cluster_case).mean())
    print('---------------------------------')
    print(pca_df_mn[cluster_case].value_counts())
    print('---------------------------------')
    viz_list = [['pc1','pc2']]
    for i, viz in enumerate(viz_list):
        plt.figure(i,(10*1.5*0.5, 7*1.2*0.5))
        colors=['red', 'green','blue','yellow','purple','black']
        ax = sns.scatterplot(x=viz[0], y=viz[1], data=pca_df_mn, palette=colors,hue=cluster_case,s=12, alpha=0.5)
        ax.set_xlabel('PC1',fontsize=13,labelpad=15)
        ax.set_ylabel('PC2',fontsize=13,labelpad=15)
        ax.legend(loc='upper left', numpoints=1, ncol=6, fontsize=13*0.7, bbox_to_anchor=(0, -0.25), title='Grid of six clusters',title_fontsize=15*0.7)
    plt.savefig("pca_viz2d.pdf", format="pdf", bbox_inches="tight")
```
